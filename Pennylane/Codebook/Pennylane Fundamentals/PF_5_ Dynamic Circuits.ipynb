## Dynamic Circuits

```python
import pennylane as qml
from pennylane import numpy as np
```

### BOOM! - bomb tester

```python
n_shots = 10000
dev = qml.device("default.qubit", shots=n_shots)
np.random.seed(0)


@qml.qnode(dev)
def circuit():
    """
    This quantum function implements the 'bomb tester' for a live bomb using mid-circuit measurements
    and returns relevant statistics with qml.counts
    """
      # 1st beam-splitter
    qml.Hadamard(wires=0)
    m_bomb =   qml.measure(wires=0, postselect=0)# measure and postselect
      # 2nd beam-splitter
    qml.Hadamard(wires=0)
    m_det =  qml.measure(wires=0) # measure (detectors)
    return  qml.counts([m_bomb, m_det]) # collect counts for m_bomb and m_det

results = circuit()
favorable_cases = results.get("00", 0)
prob_suc = favorable_cases / n_shots

print('The success probability is:', prob_suc)
```

### Improved tester

```python
n_shots = 10000
dev = qml.device("default.qubit", shots=n_shots)
np.random.seed(0)


@qml.qnode(dev)
def circuit():
    """
    This quantum function implements an improved version of 'bomb tester'
    and returns relevant statistics with qml.counts 
    """
    # copy and paste your previous code here

    # input four lines of code below
    qml.Hadamard(wires=0)                 # 1st beam-splitter
    m_bomb = qml.measure(0, postselect=0)               # measure interaction with bomb
    qml.Hadamard(wires=0)                 # 2nd beam-splitter
    m_det = qml.measure(0)                # detectors 1

    qml.cond(m_det == 1, qml.Hadamard)(wires=0)
    qml.measure(0, postselect=0)             
    qml.Hadamard(wires=0)          
    m_det_2 = qml.measure(0)
    
    return qml.counts(op=[m_bomb, m_det]), qml.counts(op=[m_det, m_det_2])

results = circuit()
print(results)
prob_suc_1 = results[0]["00"] / n_shots
prob_suc_2 = results[1]["10"] / n_shots
prob_suc = prob_suc_1 + prob_suc_2
print("The success probability is", prob_suc)
```
