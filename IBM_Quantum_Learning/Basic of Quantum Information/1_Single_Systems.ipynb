## 1. Classical information
A classic state corresponds to a finite number, or a superposition of finite numbers, e.g., binary with 0 or 1. It can also correspond to several states for a voltage (low, mid, high, off).

Representation of a probabilistic state for a quantum state:
$$ P(X = 0) = \alpha ,\space\space P(X = 1) = \beta $$

The sum of the states is 1, and the values are between 0 and 1 inclusive.
Dirac notation about the quantum state:
$$|0\rangle = \binom{\alpha}{0}, \space\space |1\rangle = \binom{0}{\beta} $$

Vectors of this form are called standard basis vectors. Each vector can be expressed as a linear combination of the basis vectors (here |0> and |1>) (called kets).

What happens if we measure a probabilistic system?\
    we obtain a classical state, and the probabilistic state becomes the measured value.
$$P(X = \alpha) = 1 $$
    
Deterministic operations:
For any function of $f: \sigma \to \sigma$ described as a deterministic operation, the transformation of a classical state $a$ into $f(a)$, for any $a$ belonging to $\sigma$. This matrix has a value equal to 1 in each column and 0 for all other entries.
$$M|a\rangle = |f(a)\rangle $$

Allows the construction of matrices, see Pauli matrix as an example.
Bra Notation:
$$\langle x| = (1 \space 0), \space \space |x\rangle = \binom{1}{0} $$

The product of a column and a row gives a matrix (example with column vectors of dimension 2).
$$ |\alpha\rangle\langle\beta| = \begin{pmatrix} *&*  \\ *&* \end{pmatrix} $$

Deterministic matrix expression, for any function $f: \sigma \to \sigma$ :
$$ M = \sum_{a \in \Sigma}^{}|f(a)\rangle\langle a| $$

Notion of Inner product.
Probabilistic operations are conventional operations that can introduce uncertainty or randomness.

## 2. Quantum Information
### A. Quantum State Vector
A quantum state is represented by a column vector that indicates the classical states it can be in. The entries are complex, and the sum of the absolute values squared is equal to 1. (Euclidean norm)
$$ v = \begin{pmatrix}
\alpha_1 \\
 ...\\
\alpha_n
\end{pmatrix}
\Rightarrow 
||v|| = \sqrt{\sum_{i=1}^{n}|\alpha_i|^2} = 1 $$
These are unit vectors. We can illustrates with an example (eg. plus/minus state):
$$|+\rangle = \frac{1}{\sqrt{2}}|0\rangle + \frac{1}{\sqrt{2}}|1\rangle,$$
$$|-\rangle = \frac{1}{\sqrt{2}}|0\rangle - \frac{1}{\sqrt{2}}|1\rangle $$

Example with complex values:
$$|\psi\rangle = \frac{1+2i}{3}|0\rangle - \frac{2}{3}|1\rangle $$

The complexe conjugate of the ket is bra (and vice versa):
$$ \langle\psi| = |\psi\rangle^\dagger $$

Examples:
$$|\psi\rangle = \frac{1+2i}{3}|0\rangle - \frac{2}{3}|1\rangle \space = \binom{\frac{1+2i}{3}}{-\frac{2}{3}} \\  $$
$$\langle\psi| = \frac{1-2i}{3}\langle0| - \frac{2}{\sqrt{3}}\langle1| \space = \begin{pmatrix}
 \frac{1-2i}{3}& {-\frac{2}{3}}
\end{pmatrix} $$


### B. Standard Basis Measurement 
Measurement of a quantum state (standard basis measurement)

- The possible outputs are classical states
- The probability of obtaining each classical state corresponds to the absolute value squared of the state vector value. For example, for the state $|+\rangle$, the probability of obtaining $0$ is $1/2$, and the probability of obtaining $1$ is $1/2$. After the measurement is performed, the state vector becomes the vector corresponding to the measured value with a probability of $1$. Therefore, the state vector changes after the measurement.

### C. Unitary Operations
A matrix U is said to be unitary if the product of U and its complex conjugate gives the identity matrix, i.e.
$$ UU^\dagger = I = U^\dagger U$$

and, there is conservation of the norm, for $v$ a state vector:
$$ \parallel Uv \parallel = \parallel v \parallel  $$

Unitary Operations on Qubits, example with Pauli Matrix : 
$$I = \begin{pmatrix} 1&0  \\ 0&1 \end{pmatrix}, \space \sigma_x = \begin{pmatrix} 0&1  \\ 1&0 \end{pmatrix} , \space \sigma_y = \begin{pmatrix} 0&-i  \\ i&0 \end{pmatrix} , \space \sigma_z = \begin{pmatrix} 1&0  \\ 0&-1 \end{pmatrix} $$

The operator $\sigma_x$ is called a bit flit, and the operator $\sigma_z$ is called a phase flit.
$$\sigma_x |0> =|1>$$
$$\sigma_x |1> =|0>$$
$$\sigma_z |0> =|0>$$
$$\sigma_z |1> =-|1> $$

Hadamard operator:
$$H = \begin{pmatrix} \frac{1}{\sqrt{2}} &\frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}}&-\frac{1}{\sqrt{2}} \end{pmatrix} $$

Useful relations with hadamard gate to know:
$$ H|0\rangle = |+\rangle $$
$$ H|1\rangle = |-\rangle $$
$$ H|+\rangle = |0\rangle $$
$$ H|-\rangle = |1\rangle $$


Phase operator, for $\theta \in \mathbb{R}$:
$P_\theta = \begin{pmatrix} 1&0  \\ 0&e^{i\theta} \end{pmatrix}$

Importants operators
S Gate, or Z90, or phase gate:
$$S = P_{\pi/2} = \begin{pmatrix} 1&0  \\ 0&i \end{pmatrix} $$

T gate is related to S gate by the equation $S = T^2$:
$$T = P_{\pi/4} = \begin{pmatrix} 1&0  \\ 0&\frac{1+i}{\sqrt{2}} \end{pmatrix} $$

Unitary matrix composition is represented as matrix multiplication. Example for square root Not $R$:
$$ R =  HSH = \begin{pmatrix} \frac{1+i}{2}&\frac{1-i}{2}  \\ \frac{1-i}{2}&\frac{1+i}{2} \end{pmatrix} $$ 
$$ R^2 = (HSH)^2 = \begin{pmatrix} 0&1  \\ 1&0 \end{pmatrix} $$

## 3. Implementation on Qiskit

```python
from qiskit import __version__
print(__version__)
```

```python
import numpy as np

# representation of ket 0 and ket 1
ket0 = np.array([[1],[0]])
ket1 = np.array([[0],[1]])

print(ket0 / 2 + ket1 / 2)
```

```python
# Creation of a matrix which represente an operation
M1 = np.array([[1, 1], [0, 0]])
M2 = np.array([[1, 0], [0, 1]])
M = M1 / 2 + M2 / 2
print(M)
```

```python
# Matrix product
print(np.matmul(M1, ket1))
print(np.matmul(M1, M2))
print(np.matmul(M, M))
```

In qiksit, we have the option to display ket and matrices with latex representation to get a better view on components. see below:

```python
from qiskit.visualization import array_to_latex

display(array_to_latex(np.matmul(M1, ket1)))
display(array_to_latex(np.matmul(M1, M2)))
display(array_to_latex(np.matmul(M, M)))
```

To initialize a quantum state vector in qiskit, the function $Statevector$ allows that:


```python
from qiskit.quantum_info import Statevector
from numpy import sqrt

# Creation of Quantum State vector on qiskit
u = Statevector([1 / sqrt(2), 1 / sqrt(2)])
v = Statevector([(1 + 2.0j) / 3, -2 / 3])
w = Statevector([1 / 3, 2 / 3])
```

```python
# Functions which allows to display numerical values, on latex format, and source code of the state
display(u.draw("text"))
display(u.draw("latex"))
print(u.draw("latex_source"))
```

```python
# Check if the State vector is a valid state
display(u.is_valid())
display(w.is_valid())
```

```python
display(v.draw("latex"))
```

```python
# Realisation of a measure on a quantum state vector, the result could be different in function of the measure
outcome, state = v.measure()
print(f"Measured: {outcome}\nPost-measurement state:")
display(state.draw("latex"))
```

```python
# To realize 1000 measures in this example to get the measures probabilities
from qiskit.visualization import plot_histogram

statistics = v.sample_counts(1000)
plot_histogram(statistics)
```
