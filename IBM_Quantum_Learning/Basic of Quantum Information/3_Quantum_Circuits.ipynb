## 2. Inner products orthonormality & projections

remind : ket and bra
$$ |\psi\rangle =
\begin{pmatrix}
\alpha_1 \\
.        \\
.        \\
\alpha_n 
\end{pmatrix} , \space \langle\psi| = \begin{pmatrix}
\bar{\alpha_1} & ... & \bar{\alpha_n} 
\end{pmatrix}$$

where $\left\{ \alpha_1, ..., \alpha_n \right\} \in \mathbb{C}^N$

$\textbf{Inner Product}$

inner product of two vector $\phi$ and $\psi$ is:
$$<\phi|\psi> =
\begin{pmatrix}
\bar{\alpha_1} & ... & \bar{\alpha_n} 
\end{pmatrix}
\begin{pmatrix}
\beta_1 \\
.        \\
.        \\
\beta_n 
\end{pmatrix} , \space  = \bar{\alpha_1}\beta_1 + ... +\bar{\alpha_n}\beta_n
$$

Reminder on the scalar product, trigonometric representation, link between the scalar product and the norm:
$$\langle\psi|\psi\rangle = \parallel |\psi\rangle\parallel^2$$

Properties of Inner product:
- Bilinearity
- Cauchy Schwartz inequality:
$$ |\langle\psi|\phi\rangle| \le  \space\parallel|\psi\rangle\parallel\times\parallel|\phi\rangle\parallel$$


$\textbf{Orthogonality and Orthonormality}$

<u>definition</u>: two vectors are orthogonal to each other if their scalar product is zero.

<u>definition</u>: A basis of vectors is said to be orthogonal if, for all vectors that are different from each other, the scalar product is zero. 

<u>definition</u>: A basis of vectors is called orthonormal if, for any two vectors that are different from each other, their scalar product is zero; otherwise, the scalar product of a vector with itself is equal to 1.

Example of orthonormal basis:

- $|+\rangle, |-\rangle$ is an orthonormal basis
- $|\phi^+\rangle,|\phi^-\rangle,|\psi^+\rangle,|\psi^-\rangle$ is an orthonormal basis

If a basis of dimension m is an orthonormal basis in an n-dimensional space, there exists a combination of n-m+1 complementary vectors to form an orthonormal basis of dimension m. See Gram-Schmidt process.

Orthonormal bases are closely related to unitary matrices:
1. The matrix U is unitary
2. The rows of U form an orthonormal basis
3. The columns of U form an orthonormal basis

$\textbf{Projections}$

A matrix $\Pi $ is called a projection if it satisfies two properties:
1. $\Pi = \Pi^\dagger$
2. $\Pi^2 = \Pi$

Measurement projection:
A collection of projections measurments $\left\{ \Pi_1,...,\Pi_m \right\}$ satisfies:
$$\Pi_1+...+\Pi_m = \mathbb{I}$$
It works because the set of projection measurements is orthogonal.\
This describes a projective measurement:
1. The output $k \in \left\{ 1,...,m \right\}$ of the measurement is chosen randomly and:
$$P(X = k) = \parallel\Pi_k|\psi\rangle\parallel^2 = \langle\psi|\Pi_k|\psi\rangle$$
2. The state become :
$$\frac{\Pi_k|\psi\rangle}{\parallel\Pi_k|\psi\rangle\parallel}$$

Example:\
Let's assume that we measure a state $\psi$:
$$|\psi\rangle = \sum_{a\in\Sigma}\alpha_a|a\rangle$$

Each output appears with a probability: 
$$\parallel|a\rangle\langle a|\psi\rangle\parallel^2=|\alpha_a|^2$$

Conditioned by output a, the state becomes:
$$\frac{|a\rangle\langle a|\psi\rangle}{\parallel|a\rangle\langle a|\psi\rangle\parallel}=\frac{\alpha_a}{|\alpha_a|}|a\rangle$$

Example of projection:
$$\Pi_0= |\phi^+\rangle\langle\phi^+|+|\phi^-\rangle\langle\phi^-|+|\psi^+\rangle\langle\psi^+|$$
$$\Pi_1= |\psi^-\rangle\langle\psi^-|$$

Any measurement projection can be implemented using unit matrices and standard measurement bases.

## 3. Limitations of Quantum informations:
#### A. Irrelevance of global phases


<u>definition</u> : Suppose that $|\psi>$ and $|\psi>$ are two states that satisfy:
$$|\phi\rangle = \alpha|\psi\rangle$$
It is said that the two states are different from a global phase. (This requires that $|\alpha|=1$ equivalent to $\alpha = e^{i\theta}$ for $\theta \in \mathbb{R}$.

Consequently, for the measurement, we obtain the same probability of measuring both states.

As a result, two quantum states that differ by a global phase are completely indistinguishable and are considered equivalent. This is equivalent to saying that there are two vector representations for the same state.

Example:\
$|-\rangle, -|-\rangle$ differ from a global phase\
$|+\rangle, |-\rangle$ differ from a relative phase because they are discernible. (example with Hadamard gate).

#### B. No-cloning Theorem
Let $X$ and $Y$ each have a set of classical states $\left\{ 0,...,d-1 \right\}$, where $d\ge2$. There is no unitary operation $U$ on the pair $(X,Y)$ such that:
$$\forall|\psi\rangle:U(|\psi\rangle\otimes|0\rangle)=|\psi\rangle\otimes|\psi\rangle$$

Consequently, there is no quantum gate or circuit that can clone a quantum state.

Notes:
1. Approximate forms of the cloning theorem are known.
2. Copying a standard basis is possible; the non-cloning theorem does not contradict this.
3. Cloning a (classical) probabilistic state is impossible.

#### C. Non orthogoanl states cannot be perfectly discriminated
It is not possible to perfectly discriminate between two non-orthogonal quantum states. By equivalence, if two quantum states can be perfectly discriminated, then they are orthogonal.

Two quantum states are perfectly discernible if there is a gate or quantum circuit capable of giving a value of 1 for a measurement of one state and a value of 0 for a measurement of the other state.

# 4. Implementation on qiskit

```python
from qiskit import __version__
print(__version__)
```

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.quantum_info import Operator
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
# Creation of a circuit with 1 qubits
circuit = QuantumCircuit(1)
circuit.h(0) #add hadamard gate to 0
circuit.s(0) #add s gate to 0
circuit.h(0) #add hadamard gate to 0
circuit.t(0) #add t gate to 0

# print the circuit
display(circuit.draw(output="mpl")) 
```

```python
# create a quantum register of 1 qubit an rename the variable "X"
X = QuantumRegister(1, "X")
circuit = QuantumCircuit(X)
circuit.h(X)
circuit.s(X)
circuit.h(X)
circuit.t(X)

display(circuit.draw(output="mpl"))
```

```python
#Creation of two registers of 1 qubit, X and Y
X = QuantumRegister(1,"X")
Y = QuantumRegister(1,"Y")

# creation of the circuit with the two registers
circuit = QuantumCircuit(Y,X)
circuit.h(Y) 
circuit.cx(Y,X)


display(circuit.draw(output="mpl")) 
display(Operator.from_circuit(circuit).draw("latex")) #get the operator from the circuit
```

```python
# Creation of the GHZ state 

#Creation of two quantum registers of 1 qubit, X and Y
X = QuantumRegister(1,"X")
Y = QuantumRegister(1,"Y")

#Creation of two classical registers of 1 bit , A and B
A = ClassicalRegister(1,"A")
B = ClassicalRegister(1,"B")

# Creation of the circuit with the different circuits
circuit = QuantumCircuit(Y,X,B,A)

circuit.h(Y)
circuit.cx(Y,X) #add CX gate
circuit.measure(Y,B) # measure of the qubit Y onto B
circuit.measure(X,A) # measure of the qubit X onto A

# Display the circuit 
display(circuit.draw(output="mpl"))
```

```python
# The circuit can be run many times to evaluate the output measures and realise an histogram
result = AerSimulator().run(circuit).result()
statistics = result.get_counts()
display(plot_histogram(statistics))

# With a GHZ state, we should get equivalent output of 00 and 11
```
