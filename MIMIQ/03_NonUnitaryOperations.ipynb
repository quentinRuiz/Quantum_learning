# Non - unitary operations
Contrary to unitary gates, non-unitary operations based on measurements make the quantum state collapse. Find in the following sections all the non-unitary operations supported by MIMIQ.
The Non unitary operations treats are:

- Measure
- Reset
- Conditional logic
- Operators

```python
# modules
from mimiqcircuits import *
import math
```

### Measure

```python
# Measure the circuit
circuit = Circuit()
circuit.push(Measure(), 0, 0) # qubit and classical register
circuit.draw()
```

```python
# measure on multiples qubits
circuit.push(Measure(), range(0,10), range(0,10))
circuit.draw()
```

```python
# reset
circuit = Circuit()
circuit.push(Reset(), 0)
circuit.draw()
```

```python
# reset is available for noise 
Reset().krausoperators()
```

```python
# create a reset and store the value in the classical register
circuit = Circuit()
circuit.push(MeasureReset(), 0,0)
circuit.draw()
```

### Conditional logic

```python
# if statement
IfStatement(GateX(), BitString("111"))
```

```python
circuit = Circuit()
circuit.push(IfStatement(GateX(), BitString("101")), 0,1,2,3)
circuit.draw()
```

### Operators
Mathematical defintion:
Operators refer to any linear operation on a state. An operator does not need to be unitary, as is the case of a gate. This means that any 
$2^N \times 2^N$ matrix can in principle represent an operator on $N$ qubits.

```python
# define operator
op = SigmaPlus()
ev = ExpectationValue(op)
```

```python
# add operator to the circuit
circuit = Circuit()
circuit.push(ev, 0, 0)
circuit.draw()
```

```python
gamma = 0.1
k1 = DiagonalOp(1, math.sqrt(1-gamma))
k2 = SigmaMinus(math.sqrt(gamma))
kraus = Kraus([k1,k2])
```

```python
gamma = 0.1
amdamp = AmplitudeDamping(gamma)
amdamp.krausoperators()
```
