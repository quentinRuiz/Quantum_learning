# Unitary Gates
This module gives the informations about how to add unitary gates to a circuit, such as :
- single qubit gates
- single qubit parametric gates
- Two qubit gates
- Two qubit parametric gates
- Multi-qubits gates
- Generalized gates
- Custom Gates


Some instruction are shown, in order to apply instruction on unitary gates such as:
- Control
- Power
- invese
- Parallel


```python
# Import module
from mimiqcircuits import *
import numpy as np
import math
```

## Unitary gates in MIMIQ

```python
# single qubit
circuit = Circuit()
circuit.push(GateX(), 0) # X gate
```

```python
# single qubit parametric gates 
circuit = Circuit()
circuit.push(GateRX(math.pi/2), 0) # RX gate
```

```python
# two qubit gates
circuit = Circuit()
circuit.push(GateCH(), 0,1) # Controlled hadamard
```

```python
# two qubit parametric gates
circuit = Circuit()
circuit.push(GateRXX(math.pi/2), 0, 1) # RXX gate
circuit.draw()
```

```python
# multi qubit gates 
circuit = Circuit()
circuit.push(GateC3X(), 0,1, 2, 3) # CCCX Gate
circuit.draw()
```

```python
# generalized gates
circuit = Circuit()
circuit.push(PhaseGradient(10), *range(0,10)) # Phase gradient on 10 qubits
circuit.draw()
```

#### Custom gates

the following instruction add a custom gate on a circuit which is :
$$ U_{custom} = \begin{pmatrix}
e^{j\pi/3} & 0 & 0 & 0 \\
0 & e^{j\pi/5} & 0 & 0 \\
0 & 0 & e^{j\pi/7} & 0 \\
0 & 0 & 0 & e^{j\pi/11} \\
\end{pmatrix} $$

```python
# only one qubit or two qubits gate can be created using MIMIQ's GateCustom
custom_matrix = np.array([[np.exp(1j*math.pi/3), 0, 0, 0], [0, np.exp(1j*math.pi/5), 0, 0 ],
                          [0, 0, np.exp(1j*math.pi/7), 0], [0, 0, 0, np.exp(1j*math.pi/11)]])
circuit = Circuit()
custom_gate = GateCustom(custom_matrix)
circuit.push(custom_gate, 0, 1)

circuit.draw()
```

## Composition
Control, Power, Inverse, Parallel instructions


```python
# add a control parameter to a choosen gate 
CX = control(1, GateX()) #example for one control qubit
circuit = Circuit()
circuit.push(CX, 0,1)
circuit.draw()
```

```python
#example for multiple control qubit
CCCCCX = control(5, GateX())

circuit = Circuit()
circuit.push(CCCCCX, *range(0,5),5)
circuit.draw()
```

```python
# power: apply power to an operator 
print(power(GateS(), 1/2))

# inverse: inverse an operator 
inv_H = inverse(GateH())
print(inv_H)

# paralell: create a composite gate applying a specific gate to multiple qubits at once you can use the {parallel()} method
circuit = Circuit()
X_gate_4  = parallel(4, GateX())
circuit.push(X_gate_4, *range(0,4))
circuit.draw()
print(X_gate_4.num_repeats)
```

```python
# be careful when using multi_qubit gate with {parallel()} as the index of the target qubits in {push()} can become confusing.
# for example see below
circuit = Circuit()
double_CX = parallel(2, GateCX())
circuit.push(double_CX, *range(0,4))

circuit.draw()
```
