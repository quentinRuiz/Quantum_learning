# Circuits
Module about the creation and instructions for a circuit with MIMIQ.\
The first example consist of the creation of the GHZ state with 4 qubits, 1 Hadamard gate and 4 CNOT gates.

```python
# importation of the module
from mimiqcircuits import *
```

```python
# Creation of the circuit (empty), use {Circuit()} instruction.
ghz = Circuit()

# add Hadamard on qubit 0
ghz.push(GateH(), 0) 
```

```python
# add CNOT gates from qubit 0 to 3
ghz.push(GateCX(), range(0,3), range(1,4)) 
```

```python
# draw the circuit, {draw()}
ghz.draw()
```

MIMIQ is a simulator, so we can get differents registers which are quantum/ classical or Z register. 
In this case, we can follow all the instructions from the begin to the end, and keep an eye on the evolution state.

```python
# empty circuit
circuit = Circuit()

# add X to the first qubit of the Quantum Register
circuit.push(GateX(), 0)

# compute Expectation vale of qubit 1 and store complex number on the first Z-register
ev = ExpectationValue(GateZ())
circuit.push(ev, 0, 0)

# Measure the qubit state and store bit into the first classical register
circuit.push(Measure(), 0,0)
```

```python
circuit.draw()
```

```python
# get the variable of the circuit 
circuit.num_qubits(), circuit.num_bits(), circuit.num_zvars()
```

```python
# to get information about something use {help(instruction)}
help(Gate)
```

```python
# To add an instruction on a circuit, the instruction {push} allow that.
circuit = Circuit()
# add X gate from qubits 0 to 10.
circuit.push(GateX(), range(0,11))
circuit.draw()
```

```python
# An other example of circuit with Id gate and CNOT gates.
circuit = Circuit()
circuit.push(GateID(), 0)

# add 3 CX gates
circuit.push(GateCX(), range(0,3), 3)

# add 3 CX gates
circuit.push(GateCX(), 0,range(1,4))

# add 3 CX gates
circuit.push(GateCX(), range(0,3), range(3,6))

circuit.draw()

# only the minimum size vector of range is took account to establish CX gates 
```

```python
# If we use of () or [], it have no influence on the instruction.
circuit = Circuit()
circuit.push(GateID(), 0)

circuit.push(GateCX(), (0,1), (2,3))

circuit.push(GateCX(), [0,2], [1,3])

circuit.draw()
```

```python
# Can insert an operation at a specific position with {insert()} instruction
circuit = Circuit()
circuit.push(GateX(), 1)

circuit.push(GateZ(), 1)

# insert the gate at a specific index
circuit.insert(2, GateY(), 1)
```

```python
# to append a circuit to another use the {append()} instruction
circuit1 = Circuit()
circuit1.push(GateX(), range(1,4))
print(circuit1)

circuit2 = Circuit()
circuit2.push(GateY(), range(1,4))
print(circuit2)


circuit1.append(circuit2)
print(circuit1)
```

```python
# To decompose the circuit, instruction {decompose()}
circuit = Circuit()
circuit.push(GateX(), 0)
circuit.decompose()
```
